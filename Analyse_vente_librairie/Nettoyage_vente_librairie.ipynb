# **Date Transactions**
Change le format de la colonne date

```python
# Je filtre sur mes date avec des valeur test
df_transactions[df_transactions['date'].str.startswith('test')]
```

```python
# Je retire donc avec str.replace()
df_transactions['date'] = df_transactions['date'].str.replace('test_', '')
# je check si les valeur sont encore la
df_transactions.loc[df_transactions['date'] == 'test_2021-03-01 02:30:02.237419']
```

```python
# je change le type de la colonne date en datetime et j'enleve les microseconde avec fonction anonyme lambda pour chaque ligne je replace(microsecond=0)
df_transactions['date'] = pd.to_datetime(df_transactions['date']).apply(lambda x: x.replace(microsecond=0))
# Je crée un colonne avec les jour mois année de mon 'date'
df_transactions['year_month_day'] = df_transactions['date'].dt.strftime('%Y-%m-%d')
df_transactions['year_buy'] = df_transactions['date'].dt.strftime('%Y')
df_transactions['Ymonth'] = df_transactions['date'].dt.strftime('%Y-%m')
df_transactions.head()
```

```python
# Je merge df_transactions et df_customers sur client id
df_transac_customer = pd.merge(df_transactions, df_customers, on=['client_id'], how='left')
print('df_customers : ',df_customers.shape)
print('df_transactions : ',df_transactions.shape)
print('df_transac_customers : ',df_transac_customer.shape)
df_transac_customer.head()
```
